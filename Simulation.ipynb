```python
from quantumnet.components import Network, Host
from quantumnet.objects import Qubit, Logger
import random
Logger.activate(Logger)  

rede = Network()
rede.set_ready_topology("Grade", 3,3)
rede.draw()
```

```python
import csv

# Função para exportar as métricas em duas colunas separadas: "Métrica" e "Valor"
def export_metrics_to_csv(metrics, file_path):
    """
    Exporta as métricas para um arquivo CSV com duas colunas: 'Métrica' e 'Valor'.
    
    Args:
        metrics: Dicionário contendo as métricas.
        file_path: O caminho do arquivo onde o CSV será salvo.
    """
    with open(file_path, mode='w', newline='') as file:
        writer = csv.writer(file)
        
        # Escreve o cabeçalho do CSV
        writer.writerow(['Métrica', 'Valor'])
        
        # Grava os dados no formato "Métrica", "Valor"
        for metric, value in metrics.items():
            writer.writerow([metric, value])  # Cada linha terá a métrica em uma coluna e o valor na outra
    
    print(f"Métricas exportadas com sucesso para {file_path}")

# Coleta as métricas
metrics = {
    "Timeslot Total": rede.get_total_timeslot(),
    "EPRs Usados": rede.get_total_useds_eprs(),
    "Qubits Usados": rede.get_total_useds_qubits(),
    "Fidelidade na Camada de Transporte": rede.transportlayer.avg_fidelity_on_transportlayer(),
    "Fidelidade na Camada de Enlace": rede.linklayer.avg_fidelity_on_linklayer(),
    "Média de Rotas": rede.networklayer.get_avg_size_routes()
}

# Exporta as métricas para o arquivo CSV
export_metrics_to_csv(metrics, r"C:\Users\polya\Downloads\QuantumNetOfficialtestmetrica1.0 (1)\QuantumNetOfficialtestmetrica\metrics_output.csv")
```
